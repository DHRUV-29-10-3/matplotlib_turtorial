```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x=np.arange(1,11)
y=np.arange(11,21)
```

```python
plt.plot(x,y,'black',linewidth=1,linestyle='dashed')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('first_plot')
plt.savefig('first_plot.png')
```

```python
"""subplot"""
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'black',linewidth=1,linestyle='dashed')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('first_plot')

plt.subplot(2,2,4)
plt.plot(x,y,'red',linewidth=1,linestyle='dashed')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('second plot')

```

```python
x=np.arange(1,11,0.05)
y=np.sin(x) 

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('graph of sinx')
```

```python
x=np.arange(1,11,0.05)

y=np.sin(x)+np.cos(x)

plt.plot(x,y)
```

```python
x=[1,2,3]
y=[10,20,30]
z=[30,20,10]

plt.subplot(2,2,1)
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.bar(x,y)

plt.subplot(2,2,2)
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.bar(x,z)
```
